```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading Dataset
data=pd.read_csv(r'C:\Users\Ayush Lal\Documents\GIT_TEST\Income-Classification\Datasets\ICclean.csv')
data
```

## Handling categorical variables with numerically coded values

```python
data['sex'] = data['sex'].map({' Male': 0, ' Female': 1}).astype(int)
```

```python
data['race'] = data['race'].map({' Black': 0, ' Asian-Pac-Islander': 1,' Other': 2, ' White': 3, ' Amer-Indian-Eskimo': 4}).astype(int)
```

```python
data['workclass'] = data['workclass'].map({' Self-emp-inc': 0, ' State-gov': 1,' Federal-gov': 2, ' Without-pay': 3, ' Local-gov': 4,' Private': 5, ' Self-emp-not-inc': 6}).astype(int)
```

```python
data['education'] = data['education'].map({' Some-college': 0, ' Preschool': 1, ' 5th-6th': 2, ' HS-grad': 3, ' Masters': 4, ' 12th': 5, ' 7th-8th': 6, ' Prof-school': 7,' 1st-4th': 8, ' Assoc-acdm': 9, ' Doctorate': 10, ' 11th': 11,' Bachelors': 12, ' 10th': 13,' Assoc-voc': 14,' 9th': 15}).astype(int)
```

```python
data['occupation'] = data['occupation'].map({ ' Farming-fishing': 1, ' Tech-support': 2,' Adm-clerical': 3, ' Handlers-cleaners': 4, 
 ' Prof-specialty': 5,' Machine-op-inspct': 6, ' Exec-managerial': 7,' Priv-house-serv': 8,' Craft-repair': 9,' Sales': 10, ' Transport-moving': 11, ' Armed-Forces': 12, ' Other-service': 13,' Protective-serv':14}).astype(int)
```

```python
data.drop(columns=['education-num','marital-status','relationship','native-country'],inplace=True)
data.head()
```

## Perform univariate and bivariate analysis and derive meaningful insights about the dataset

```python
sns.set_style('darkgrid')
```

### 1) Age 
#### Univariate

```python
d=data.groupby('age')["age"].count()
sns.scatterplot(x=d.index,y=d)
plt.xlabel("Age")
plt.ylabel("No. of Entries")
plt.title("Age VS Entries Graph")
plt.show()
```

### shows the age distribution among the entries in our dataset. The ages range from 17 to 90 years old with the majority of entries between the ages of 25 and 50 years.


#### Bivariate

```python
g = sns.FacetGrid(data, col='income')
g = g.map(sns.distplot, "age")
plt.show()
```

### Looking at the graph, we can see that there is a The most interesting ratios to note are those of groups 17-20,71-80, and 81-90 where there is almost no chance to have an income of greater than 50K. 


### Work Class

#### Univariate

```python
d=data.groupby('workclass')["workclass"].count()
sns.barplot(d,d.index,orient='h',ci=0,palette='Spectral')
plt.ylabel("Work Classes")
plt.xlabel("No. of Entries")
plt.title("Work Classes VS Entries Graph")
plt.yticks(np.arange(7),('Self-emp-inc', ' State-gov',' Federal-gov',' Without-pay', ' Local-gov',' Private', ' Self-emp-not-inc'))
plt.show()
```

### The majority of the individuals work in the private sector.

#### Bivariate

```python
sns.barplot(x='workclass',y='income',data=data,ci=0)
plt.ylabel("Probability that income is >=50 K")
plt.title('Work Class VS Income')
plt.xticks(np.arange(7),('Self-emp-inc', ' State-gov',' Federal-gov',' Without-pay', ' Local-gov',' Private', ' Self-emp-not-inc'),rotation=90)
plt.show()
```

### The probabilities of making above 50K are similar among the work classes except for selfempinc and federal government. which has more probability of  earning more than  50K

```python
from scipy import stats
from functools import reduce
x=np.abs(stats.zscore(data['age']))
y=np.abs(stats.zscore(data['workclass']))
threshold = 3
b=np.where(x > 3)
c=np.where(y > 3)
out=reduce(np.union1d,(b,c))
```
